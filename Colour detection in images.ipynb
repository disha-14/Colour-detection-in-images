## Task 2 : Colour Identification in Images

For this task I have used OpenCV in Python to identify/detect various colours in a given image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Now that we have imported the required libraries, we can display the image we want. We can also find out the shape and dimensions of the image.

```python
image = cv2.imread("Downloads/sunflower.jpg")
print("Shape: {}".format(image.shape))
plt.imshow(image)
```

Here we see that the colours seen in the picture dont seem correct. This is because OpenCV takes BGR as default format instead of RGB. So we convert it into RGB format by doing the following-

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
```

We can also convert this coloured image to gray if we want by doing the following

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

To detect different colours, OpenCV uses HSV format. So we convert the image into HSV format in order to detect the colours we want.

```python
HSV_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

Now, let's say we want to detect the yellow colour in the above image, we just need 2 arrays where the parameters passed are Hue, Saturation and Value respectively. Each colour has a range of values and depending upon this range we can detect the colour.

```python
lower = np.array([25, 150, 50])
upper = np.array([35, 255, 255])
```

We mask the given image so that only the regions where the colour is present is masked and the rest of the image is turned to black colour

```python
mask = cv2.inRange(HSV_image, lower, upper)
plt.imshow(mask)
```

```python
final = cv2.bitwise_and(RGB_image, RGB_image, mask = mask)
plt.imshow(final)
```
